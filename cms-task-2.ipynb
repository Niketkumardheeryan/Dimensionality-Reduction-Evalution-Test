### Given Values

```python
e = 4.80 *(10** -10)
f = 0.4164
n = 0.1
x = 0.1
g = 2

m = 9.11 * (10** -28)
c = 3.00 * (10** 10)
z = 2.00

t = 6.265 * (10**8)
v0 = 2.46607 * (10** 15)
```

- I(lada)= exp(-a * d)
- v=c/lamda

- a=(e2 *f *n (1-x) g) / (4 *pi * m* C *z) * T / (v-v0)^2 +(T/4 *pi) ^2



```python
# Get the values for a 

def plot_function(e,f,n,x,g,m,c,z,t,v0,d):
    
    A_list=[]

    for lmda in np.arange(91* (10** -7),150 *(10**-7),0.50 *(10**-7)):

        v=c/lmda
        a=((e**2 * f * n * (1-x) *g ) * t) / ((4 * np.pi* m*c*z ) * ((v-v0)**2 + ((t/4) * np.pi)**2))
        A_list.append(a)


    wave_len=[lmda for lmda in np.arange(91* (10** -7),150 *(10**-7),0.50 *(10**-7))]


    Intensity=[]

    for i in range(len(A_list)):
        var=np.exp(-1 * A_list[i] * d )
        Intensity.append(var)
    
    
    return wave_len,Intensity
```

```python
wave_len,intensity = plot_function(e,f,n,x,g,m,c,z,t,v0,10**14)

plt.plot(wave_len,intensity)
```
